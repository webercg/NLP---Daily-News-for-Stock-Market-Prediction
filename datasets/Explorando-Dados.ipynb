```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
arq = 'https://raw.githubusercontent.com/webercg/NLP---Daily-News-for-Stock-Market-Prediction/main/datasets/dataset-2021/01_GoogleNews_Petr_Jan_21.csv'
df_news = pd.read_csv(arq, sep='|')
```

```python
df_petro = pd.read_csv('Hist_Preço_Petr_2021_.csv', sep='|')
```

```python
def ajusta(data):
  x = data.split("-")
  y = x[2] + "/" + x[1] + "/" + x[0]
  return y
df_petro['data'] = df_petro['Date'].apply(ajusta)
```

```python
from datetime import datetime, timedelta

def atribuiData(data):
  if df_petro[(df_petro['data'] == data)].shape[0] > 0:
    data = data
  else:
      for i in range(10):
        x = datetime.strptime(data, '%d/%m/%Y').date() + timedelta(days=i)
        data_x = str( x.strftime('%d/%m/%Y') )
        if df_petro[(df_petro['data'] == data_x)].shape[0] > 0:
          data = data_x
          break
  return data
```

```python
df_news['data_influencia'] = df_news['date'].apply(atribuiData)
```
