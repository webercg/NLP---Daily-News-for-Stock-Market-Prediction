## Análise - Base 01

```python
# Carregando dataset kaggle em GitHub
dataset = "https://raw.githubusercontent.com/webercg/NLP---Daily-News-for-Stock-Market-Prediction/main/datasets/proposta-kaggle/RedditNews.csv"
df = pd.read_csv(dataset)
display(df)
```

```python
df.info()
```

```python
df.rename(columns={'Date': 'date', 'News': 'news'}, inplace = True)
```

```python
df.head(2)
```

```python
df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d')
df['ano'] = df['date'].dt.year
df.head(2)
```

```python
bar = df['ano'].value_counts().sort_index()
bar
```

```python
bar.plot(kind="bar")
```

```python
df.head(50)
```

```python
df2 = df.query('news.str.contains("finance") and news.str.contains("Brazil")', engine='python')
```

```python
display(df2)
```

```python
for i in df2.index.array:
    print(i,' - ',df2.at[i, 'news'])
```
